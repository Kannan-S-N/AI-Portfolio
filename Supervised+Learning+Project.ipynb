### Part 1.4 a - Split data into X and Y.

```python
X = combined_df.drop(labels='Class', axis=1, inplace=False)
y = combined_df['Class']
```

### Part 1.4 b - Split data into train and test with 80:20 proportion.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=365)
```

### Part 1.4 c - Train a Supervised Learning Classification base model using KNN classifier.

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
clf = KNeighborsClassifier()
clf.fit(X_train, y_train)
```

### Part 1.4 d - Print all the possible performance metrics for both train and test data.

```python
## Accuracy Score
print('Accuracy score on train data: ', clf.score(X_train, y_train).round(3))
print('Accuracy score on test data: ', clf.score(X_test, y_test).round(3))
```

```python
## Check for class imbalance
combined_df['Class'].value_counts(normalize=True)
```

From the above, we can see class imbalance in our dataset. Hence, we cannot rely solely on accuracy metric.

```python
## Prediction
y_pred = clf.predict(X_test)
y_pred
```

```python
## Confusion Matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
## Plot Confusion Matrix
sns.reset_orig() ## to remove the white lines
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(clf, X_test, y_test, cmap='magma')
```

```python
## Print Classification report
from sklearn.metrics import classification_report
print('Classification Report:  \n', classification_report(y_test, y_pred, target_names=clf.classes_))
```

### Part 1.5 a - Experiment with various parameters to improve performance of the base model.

```python
## Find the optimal value of K by plotting Error Rate vs K-Value
error_rate = []

for i in range(1,50):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    y_pred_i = knn.predict(X_test)
    error_rate.append(np.mean(y_pred_i != y_test))
```

```python
plt.figure(figsize=(20,8))
plt.plot(range(1,50),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value', fontsize=20)
plt.xlabel('K Value', fontsize=15)
plt.ylabel('Error Rate', fontsize=15)
```

From the plot we can observe that at k=15, the error rate is minimum and it increases gradually after that. So we will consider the optimal value of K=15.

```python
## New Model initialization 'clf1' using optimal K-value
clf1 = KNeighborsClassifier(n_neighbors=15)
clf1.fit(X_train, y_train)
```

```python
## Accuracy Score of model 'clf1' using optimal K-value
print('Accuracy score on train data: ', clf1.score(X_train, y_train).round(3))
print('Accuracy score on test data: ', clf1.score(X_test, y_test).round(3))
```

Accuracy score on test data has increased with the new model.

```python
## Plot Confusion Matrix of the model 'clf1' which uses optimal K-value

y_pred1 = clf1.predict(X_test) ## prediction
sns.reset_orig() ## to remove the white lines
plot_confusion_matrix(clf1, X_test, y_test, cmap='magma')
```

Number of True positives has increased and there is no misclassification observed for Class 'Type_S'. 

```python
## Print Classification report of the model 'clf1' which uses optimal K-value
print('Classification Report:  \n', classification_report(y_test, y_pred1, target_names=clf1.classes_))
```

Accuracy, f1-score, precision score, recall score have increased with the model 'clf1' using optimal K-value. Hence the performance of our model 'clf1' is far better as compared to our base model 'clf'.

```python
## New model initialization using weights='distance'
clf2 = KNeighborsClassifier(n_neighbors=15, weights='distance')
clf2.fit(X_train, y_train)
```

```python
## Confusion matrix and Classification report for new model 'clf2'
y_pred2 = clf2.predict(X_test)
plot_confusion_matrix(clf2, X_test, y_test, cmap='magma')
print('\n')
print('Classification Report:  \n', classification_report(y_test, y_pred2, target_names=clf2.classes_))
```

No further improvement is seen after initializing the Model 'clf2' with weights parameter as 'distance'. Performance of 'clf2' remains same as 'clf1'.

### Part 2.1 c - Merge both the Dataframes on ‘ID’ feature to form a single DataFrame

```python
data = pd.merge(data1,data2,how='outer',on='ID')
data
```

### Part 2.1 d - Change Datatype of below features to ‘Object’
‘CreditCard’, ‘InternetBanking’, ‘FixedDepositAccount’, ‘Security’, ‘Level’, ‘HiddenScore’ 
[Reason behind performing this operation:- Values in these features are binary i.e. 1/0. But DataType is ‘int’/’float’ which is not expected.]

### Part 2.2 a - Visualize distribution of Target variable ‘LoanOnCard’ and clearly share insights

```python
sns.countplot(x=data['LoanOnCard'])
```

We can clearly see imbalanced class distribution from the above plot. Count is very high for Class'0', while it is very less for Class'1'.

### Part 2.2 b - Check the percentage of missing values and impute if required

```python
percent_mv = data.isna().mean()*100
percent_mv
```

As the percentage of missing values in target variable 'LoanOnCard' is just 0.4%, we decide to impute the missing values with mode.

```python
data['LoanOnCard'] = data['LoanOnCard'].fillna(data['LoanOnCard'].mode()[0])
data.head()
```

```python
## Again check for missing values
data.isna().sum().sum()
```

So, there are no missing values. All missing values in 'LoanOnCard' were replaced with mode.

### Part 2.3 a- Split data into X and Y.
[Recommended to drop ID & ZipCode. LoanOnCard is target Variable]

```python
inputs = data.drop(labels=['ID', 'ZipCode', 'LoanOnCard'], axis=1)
target = data['LoanOnCard']
```

### Part 2.3 b - Split data into train and test. Keep 25% data reserved for testing.

```python
from sklearn.model_selection import train_test_split

inputs_train, inputs_test, target_train, target_test = train_test_split(inputs, target, test_size=0.25, random_state=365)
```

### Part 2.3 c - Train a Supervised Learning Classification base model - Logistic Regression.

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter=1056) ## with defualt max_iter, convergence warning appears
lr.fit(inputs_train, target_train)
```

### Part 2.3 d - Print evaluation metrics for the model and clearly share insights.

```python
from sklearn import metrics

target_pred = lr.predict(inputs_test)
print('Confusion metrics: \n', metrics.confusion_matrix(target_test, target_pred))
print('Classification Report:  \n', metrics.classification_report(target_test, target_pred))
print('ROC_AUC score:', metrics.roc_auc_score(target_test, target_pred).round(2))
```

Though the accuracy is 95%, the recall of minority class is very less. So we can say that our model is biased towards majority class.

### Part 2.3 e - Balance the data using the right balancing technique.

i. Check distribution of the target variable
ii. Say output is class A : 20% and class B : 80%
iii. Here you need to balance the target variable as 50:50.
iv. Try appropriate method to achieve the same.

```python
## check distribution of target variable
percent_dist_tv = (data.LoanOnCard.value_counts()/data.LoanOnCard.value_counts().sum())*100
percent_dist_tv
```

We can see imbalanced distribution of classes in the target variable. 90.4% belong to class'0.0' and 9.6% belong to class'1.0'. To overcome this problem we will use oversampling technique called SMOTE.

```python
## balance the target variable as 50:50
from imblearn.over_sampling import SMOTE
smote = SMOTE()
inputs_train_res, target_train_res = smote.fit_resample(inputs_train, target_train)
```

```python
## check distribution of target variable on train data
target_train_res.value_counts()
```

Thus, we have balanced the target variable as 50:50.

### Part 2.3 f- Again train the same previous model on balanced data.

```python
lr.fit(inputs_train_res, target_train_res)
```

### Part 2.3 g- Print evaluation metrics and clearly share differences observed.

```python
target_pred1 = lr.predict(inputs_test)
print('Confusion metrics: \n', metrics.confusion_matrix(target_test, target_pred1))
print('Classification Report:  \n', metrics.classification_report(target_test, target_pred1))
print('ROC_AUC score:', metrics.roc_auc_score(target_test, target_pred1).round(2))
```

The accuracy has decreased a little bit, but the recall score has increased for the minority class. ROC_AUC score has improved from 0.79 to 0.89.

### Part 2.4 a- Train a base model each for SVM, KNN.

```python
## SVM base model
from sklearn.svm import SVC

svm_base_model = SVC()
svm_base_model.fit(inputs_train, target_train)
```

```python
## KNN base model
from sklearn.neighbors import KNeighborsClassifier

knn_base_model = KNeighborsClassifier()
knn_base_model.fit(inputs_train, target_train)
```

### Part 2.4 b - Tune parameters for each of the models wherever required and finalize a model.

```python
## Tune parameters for svm model with default kernel 'rbf' using different values of 'C' and 'gamma'
c_values = []
gamma_values = []
accuracy = []
precision = []
recall = []
f1_score = []

C = [0.001,0.01,0.1,1,100,1000]
gamma = [0.001,0.01,0.1,1,100,1000]
for i in C:
    for j in gamma:
        svm_model = SVC(C=i, gamma=j)
        svm_model.fit(inputs_train, target_train)
        pred = svm_model.predict(inputs_test)
        c_values.append(i)
        gamma_values.append(j)
        accuracy.append(metrics.accuracy_score(target_test,pred))
        precision.append(metrics.precision_score(target_test,pred, zero_division=0))
        recall.append(metrics.recall_score(target_test,pred))
        f1_score.append(metrics.f1_score(target_test,pred))
```

```python
## Create a score table with different values of 'C' and 'gamma'
d = {'C': c_values, 'Gamma': gamma_values, 'Accuracy': accuracy, 'Precision': precision, 'Recall': recall, 'f1-score': f1_score}
pd.DataFrame(data=d)
```

From the above dataframe, we can say that the SVM model with C=1000 and gamma=0.001 gives the highest accuracy, precision, recall and f1 score.

```python
## Tune parameters of the knn model
weights_param = []
n_neighbors_param = []
accuracy_knn = []
precision_knn = []
recall_knn = []
f1_score_knn = []

weights = ['uniform', 'distance']
n_neighbors = [num for num in list(range(1,40)) if num % 2 != 0]

for w in weights:
    for n in n_neighbors:
        knn_model = KNeighborsClassifier(n_neighbors=n, weights=w)
        knn_model.fit(inputs_train, target_train)
        pred_knn = knn_model.predict(inputs_test)
        weights_param.append(w)
        n_neighbors_param.append(n)
        accuracy_knn.append(metrics.accuracy_score(target_test,pred_knn))
        precision_knn.append(metrics.precision_score(target_test,pred_knn))
        recall_knn.append(metrics.recall_score(target_test,pred_knn))
        f1_score_knn.append(metrics.f1_score(target_test,pred_knn))
```

```python
## Create a score table with different values of 'Weights' and 'K'
p = {'Weights': weights_param, 'K-Value': n_neighbors_param, 'Accuracy': accuracy_knn, 'Precision': precision_knn, 'Recall': recall_knn, 'f1-score': f1_score_knn}
pd.DataFrame(data=p)
```

From the above dataframe, we can say that the KNN model with weights='distance' and n_neighbors=3 gives the highest accuracy, precision, recall and f1 score.

On comparing the SVM and KNN model, the SVM model is the best as it gives the highest accuracy, precision, recall and f1 score. Hence our final model is the SVM model with C=1000 and gamma=0.001.

### Part 2.4 c- Print evaluation metrics for final model.

```python
svm_final = SVC(C=1000, gamma=0.001)
svm_final.fit(inputs_train, target_train)
pred_final = svm_final.predict(inputs_test)
print('Confusion Matrix of the final model: \n', metrics.confusion_matrix(target_test, pred_final))
print('Classification Report of the final model: \n', metrics.classification_report(target_test, pred_final))
print('ROC_AUC Score of the final model:', metrics.roc_auc_score(target_test, pred_final).round(2))
```
