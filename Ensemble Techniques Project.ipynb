### Part 1 - 2 e.Encode all the appropriate Categorical features with the best suitable approach

```python
churn_dummies = pd.get_dummies(churn, columns=['gender', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod'])
churn_dummies['Churn'] = churn_dummies['Churn'].map({'No':0, 'Yes':1})
churn_dummies
```

### Part 1- 2 f.Split the data into 80% train and 20% test.

```python
X = churn_dummies.drop(['customerID','Churn'], axis=1)
y = churn_dummies['Churn']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.20, random_state=1)
```

### Part 1 - 2 g.Normalize/Standardize the data with the best suitable approach

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train.loc[:,['tenure', 'MonthlyCharges', 'TotalCharges']] = scaler.fit_transform(X_train.loc[:,['tenure', 'MonthlyCharges', 'TotalCharges']])
X_test.loc[:,['tenure', 'MonthlyCharges', 'TotalCharges']] = scaler.transform(X_test.loc[:,['tenure', 'MonthlyCharges', 'TotalCharges']])
```

### Part 1 - 3 a. Train a model using XGBoost. Also print best performing parameters along with train and test performance.

```python
## fit model on training data
from xgboost import XGBClassifier
model = XGBClassifier(use_label_encoder=False, objective='multi:softmax', num_class=2, eval_metric='mlogloss')
model.fit(X_train, y_train)
```

```python
## make predictions for test data
y_pred = model.predict(X_test)
y_pred
```

```python
## print accuracy score
print('Accuracy score on train data: ', model.score(X_train,y_train))
print('Accuracy score on test data: ', model.score(X_test,y_test))
```

Model's performance on train data is far better than the test data. Hence, our model is overfit.

```python
## Tune the model using different learning rates

train_data_acc = []
test_data_acc = []

learning_rates = [0.05,0.075,0.1,0.125,0.15,0.175,0.2,0.225,0.25,0.275,0.3]
for i in learning_rates:
    xgb = XGBClassifier(use_label_encoder=False, objective='multi:softmax', num_class=2, eval_metric='mlogloss',learning_rate = i)
    xgb.fit(X_train, y_train)
    y_pred_i = xgb.predict(X_test)
    train_data_acc.append(xgb.score(X_train,y_train))
    test_data_acc.append(xgb.score(X_test,y_test))
```

```python
## Accuracy scores for different learning rates

d = {'Learning_Rate':learning_rates, 'Train_data_Accuracy': train_data_acc, 'Test_data_Accuracy': test_data_acc}
accuracy_table = pd.DataFrame(data=d)
accuracy_table['Diff'] = accuracy_table['Train_data_Accuracy'] - accuracy_table['Test_data_Accuracy']
accuracy_table
```

From the above table, we can conclude that the model with learning rate 0.05 is the best model.

### Part 2 - Build a machine learning pipeline that will run autonomously with the csv file and return bestperforming model.

```python
## create a function to read csv
def CSV(filename):
    return pd.read_csv(filename)    
```

```python
## create a fuction to detect null values
def missing_value_detection(dataframe):
    return dataframe.isna().sum()
```

```python
## create a fuction to remove rows with null values
def remove_null_values(dataframe):
    return dataframe.dropna(axis=0, how='any')
```

```python
## create a function to check unique values
def unique_values(dataframe):
    for i in dataframe.columns:
        print(i,':', dataframe[i].unique())
```

```python
## create data preprocessing function
def preprocessing(filename):
    dataframe = CSV(filename)
    missing_value_detection(dataframe)
    remove_null_values(dataframe)
    unique_values(dataframe)
    return dataframe
```

```python
## Import various ml algorithms

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
#Import metric for performance evaluation
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
#Defining the modelling function
def modeling(alg, alg_name):
    model = alg() 
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    #Performance evaluation
    def print_scores(alg, y_true, y_pred):
        print(alg_name)
        acc_score = accuracy_score(y_true, y_pred)
        print("accuracy: ",acc_score)
        pre_score = precision_score(y_true, y_pred)
        print("precision: ",pre_score)
        rec_score = recall_score(y_true, y_pred)                            
        print("recall: ",rec_score)
        f_score = f1_score(y_true, y_pred, average='weighted')
        print("f1_score: ",f_score)

    print_scores(alg, y_test, y_pred)
    return model
```

```python
lr_model = modeling(LogisticRegression, 'Logistic Regression')
```

```python
svm_model = modeling(SVC, 'Support Vector Machine')
```

```python
rf_model = modeling(RandomForestClassifier, 'Random Forest')
```

```python
dt_model = modeling(DecisionTreeClassifier, 'Decision Tree')
```

```python
nb_model = modeling(GaussianNB, 'Naive Bayes')
```

SVC model is the best performing model with accuracy score 0.81.
