#### Part 1 Q 2 A Probability that none of the items are defective?

```python
import numpy as np
import scipy.stats as stats
```

```python
n = 10
```

```python
p = 0.05
```

```python
k = np.arange(0,11)
k
```

```python
binomial = stats.binom.pmf(k,n,p)
binomial
```

```python
print('P(0) = ', binomial[0].round(3))
```

#### Part 1 Q 2 B Probability that exactly one of the items is defective?

```python
print('P(1) = ', binomial[1].round(3))
```

#### Part 1 Q 2 C Probability that two or fewer of the items are defective?

```python
print('P(2 or less) = ', (binomial[0]+binomial[1]+binomial[2]).round(3))
```

#### Part 1 Q 2 D Probability that three or more of the items are defective ?

```python
print('P(3 or more) = ', (1 - (binomial[0]+binomial[1]+binomial[2])).round(3))
```

#### Part 1 Q 3 A What is Probability that in a given week he will sell some cars?

```python
## 'sell some cars' means 'sell 1 or more cars'
rate = 3
n = np.arange(0,6)
```

```python
poisson = stats.poisson.pmf(n,rate)
poisson
```

```python
print('P(sell 1 or more cars) = ', (1 - poisson[0]).round(3))
```

#### Part 1 Q 3 B What is Probability that in a given week he will sell 2 or more but less than 5 cars?

```python
print('P(sell 2 or more but less than 5) = ', (poisson[2]+poisson[3]+poisson[4]).round(3))
```

#### Part 1 Q 3 C Plot the poisson distribution function for cumulative probability of cars sold per-week vs number of cars sold per week

```python
cumm_prob = stats.poisson.cdf(n,rate)
cumm_prob
```

```python
import matplotlib.pyplot as plt
plt.plot(n,cumm_prob,'o-')
plt.title('Poisson: $\lambda$ = %i ' % rate)
plt.xlabel('Number of Cars sold per week')
plt.ylabel('Cumulative Probability of cars sold per week')
plt.show()
```

#### Part 3 Q4 A Is there any significant difference between Funds raised by companies that are still operating vs companies that closed down?

```python
print('Funds raised by companies that are still operating: ', df[df['OperatingState'] == 'Operating']['Funds_in_million'].sum())
print('Funds raised by companies that are closed down: ', df[df['OperatingState'] == 'Closed']['Funds_in_million'].sum())
```

#### Part 3 Q4 B Write the null hypothesis and alternative hypothesis

The null hypothesis is that there is no difference between Funds raised by companies that are still operating vs companies that closed down.
H0: µ1 = µ2.
The alternative hypothesis is that there is difference between Funds raised by companies that are still operating vs companies that closed down.
H1: µ1 is not equal to µ2.

#### Part 3 Q4 C Test for significance and conclusion

```python
## Create two groups
group1 = df[df['OperatingState'] == 'Operating']['Funds_in_million']
group2 = df[df['OperatingState'] == 'Closed']['Funds_in_million']
```

```python
## Check variance for each group
print(np.var(group1), np.var(group2))
```

group1 and group2 are wihout equal variance.

```python
## Perform two sample t-test without equal variance
import scipy.stats as st
st.ttest_ind(a=group1,b=group2,equal_var=False)
```

Because the p-value of our test is less than alpha = 0.05, we reject the null hypothesis. Results are statistically significant.

#### Part 3 Q4 G Write your hypothesis comparing the proportion of companies that are operating between winners and contestants:

Null Hypothesis(H0) for the test is that the proportions are the same.
H0: p1 = p2. 
Alternative Hypothesis(H1) for the test is that the proportions are not the same.
H1: p1 not equal to p2

#### Part 3 Q4 H Test for significance and conclusion

```python
p1 = b/a
p2 = d/c
print(p1, p2)
```

```python
from statsmodels.stats.proportion import proportions_ztest

stat, pval = proportions_ztest([b, d] , [a, c])
```

```python
print('p-value: ', pval)
```

Since this p-value is not less than our significance level α = 0.05, we fail to reject the null hypothesis. We do not have sufficient evidence to say that the proportion of companies that are operating between winners and contestants is different.
